###2. Explain the concept of function arguments and parameters in Python.

Parameters are the variable names listed in the function definition, acting as placeholders. Arguments are the actual values passed into the function when it is called. Parameters receive arguments, allowing the function to work dynamically with different inputs.

```python
def add(a, b):
    return a + b

add(5, 3)
```

###4. What is the purpose of the return statement in a Python function?

The return statement is used to send back a value from a function to the place where it was called. It also immediately ends the function execution. Without a return, a function returns None by default.

```python
def area(r):
    return 3.14 * r * r
```

###5. What are iterators in Python and how do they differ from iterables?

An iterable is any object that can be looped over (like lists, tuples, strings), while an iterator is an object that keeps track of its position and gives the next value using __next__() or next(). Iterators are created using the iter() function.

```python
it = iter([1, 2, 3])
next(it)
```

###6. Explain the concept of generators in Python and how they are defined.

Generators are a type of iterator written using a normal function but with yield instead of return. Each time the generator function is called, it resumes from where it left off. This is memory-efficient and useful for large data streams.

```python
def numbers():
    yield 1
    yield 2
    yield 3
```

###7. What are the advantages of using generators over regular functions?

Generators are efficient because they don’t store all values in memory — they produce values one at a time. They’re also faster for large datasets, pause and resume automatically, and are ideal for infinite sequences or data pipelines.

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1
```

###8. What is a lambda function in Python and when is it typically used?

A lambda function is an anonymous, one-line function that doesn’t require a name. It's mostly used for short operations like filtering or mapping, especially when we don’t want to define a full def function for a small task.

```python
double = lambda x: x * 2
double(6)
```

###9. Explain the purpose and usage of the map() function in Python.

The map() function applies a given function to each item in a list (or any iterable) and returns a new map object. It’s commonly used to perform transformations without writing loops.


```python
list(map(lambda x: x*2, [1, 2, 3]))  # ➝ [2, 4, 6]
```

###Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.


```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for i in range(1, 201):
    if is_prime(i):
        print(i, end=" ")
```

###5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
```

###6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i
```

###7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()
```

###11. Question:
Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:


[Order Number, Book Title and Author, Quantity, Price per Item]

[34587, "Learning Python, Mark Lutz", 4, 40.95]  
[98762, "Programming Python, Mark Lutz", 5, 56.80]  
[77226, "Head First Python, Paul Barry", 3, 32.95]  
[88112, "Einführung in Python3, Bernd Klein", 3, 24.99]

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10 €, if the value of the order is smaller than 100 €.




```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

final_orders = []

for order in orders:
    order_num = order[0]
    quantity = order[2]
    price = order[3]
    total = quantity * price
    if total < 100:
        total += 10
    final_orders.append((order_num, round(total, 2)))

print(final_orders)
```
